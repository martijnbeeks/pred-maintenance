### Step 1: Select numerical and categorical data

```python
num_cols_switch = ['region_code', 'district_code']
cols_drop = ['gps_height', 'longitude', 'latitude', 'construction_year']
remove_cat_columns = ['wpt_name', 'installer', 'funder', 'subvillage', 'ward', 'scheme_name']


def get_feature_types(df_set, num_cols_switch=num_cols_switch, cols_drop=cols_drop, remove_cat_columns=remove_cat_columns):
    numerical_cols = list(df_set._get_numeric_data().columns)
    categorical_cols = list(set(df_set.columns) - set(numerical_cols))
    
    numerical_cols = [x for x in numerical_cols if x not in num_cols_switch]
    categorical_cols.extend(num_cols_switch)
    
    numerical_cols = [x for x in numerical_cols if x not in cols_drop]
    
    for drop_column in remove_cat_columns:
        if drop_column in categorical_cols:
            categorical_cols = [x for x in categorical_cols if x not in drop_column]
            
    # Drop columns with more than 10% missing values
    missing_counts = df_set.isnull().sum() / len(df_set)
    d = {k:v for (k,v) in missing_counts.items() if v>0.1}
    
    for drop_column in d.keys():
        if drop_column in categorical_cols:
            categorical_cols = [x for x in categorical_cols if x not in drop_column]
        elif drop_column in numerical_cols:
            numerical_cols = [x for x in numerical_cols if x not in drop_column]
            
    return categorical_cols, numerical_cols
    
```

```python
# Get numerical and categorical variables
categorical_cols, numerical_cols = get_feature_types(df_set)
```

### Step 3: Clean data

```python
# Separate columns in numerical and categorical
categorical_cols, numerical_cols = get_feature_types(df_set)

# Transform string labels to integers
label_map = {'non functional': 0, 'functional needs repair':1, 'functional':2}
df_labels['status_group'] = df_labels['status_group'].map(label_map)

class RecordedDate(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit( self, X, y = None ):
        return self
    
    def transform(self, X, y=None):
        X["year_recording"] = pd.to_datetime(X['date_recorded']).dt.year
        X["month_recording"] = pd.to_datetime(X['date_recorded']).dt.month
        X["day_recording"] = pd.to_datetime(X['date_recorded']).dt.day
        return X
    def get_feature_names(self):
        return ['year_recording', 'month_recording', 'day_recording']

def preprocessing_pipeline_shap(clf, categorical, numerical, scaler=StandardScaler(), encoder=OneHotEncoder(sparse=False, handle_unknown='ignore')):
    """ Returns a pipeline that imputes all missing values, encodes categorical features and scales numeric ones
    Keyword arguments:
    categorical -- A list of categorical column names. Example: ['gender', 'country'].
    scaler -- any scikit-learn feature scaling method (Optional)
    encoder -- any scikit-learn category encoding method (Optional)
    Returns: a scikit-learn pipeline which preprocesses the data and then runs the classifier
    """
    
    preprocessing_feat = Pipeline(steps=[
        ("feature_engineering", RecordedDate())
        ])
    
    categorical.extend(RecordedDate().get_feature_names())
    categorical.remove('date_recorded')
    
    preprocessing_cat = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encode', encoder)])
    
    preprocessing_num = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', scaler)])
    
    
    
    processing = ColumnTransformer(transformers=[
     ('cat', preprocessing_cat, categorical),
     ('num', preprocessing_num, numerical)
     ]
    )

    pipe = Pipeline(steps=[('feat', preprocessing_feat), 
                       ('preprocessor', processing)])
    return pipe
    

def preprocessing_pipeline(clf, categorical, numerical, scaler=StandardScaler(), encoder=OneHotEncoder(sparse=False, handle_unknown='ignore')):
    """ Returns a pipeline that imputes all missing values, encodes categorical features and scales numeric ones
    Keyword arguments:
    categorical -- A list of categorical column names. Example: ['gender', 'country'].
    scaler -- any scikit-learn feature scaling method (Optional)
    encoder -- any scikit-learn category encoding method (Optional)
    Returns: a scikit-learn pipeline which preprocesses the data and then runs the classifier
    """
    
    preprocessing_feat = Pipeline(steps=[
        ("feature_engineering", RecordedDate())
        ])
    
    categorical.extend(RecordedDate().get_feature_names())
    categorical.remove('date_recorded')
    
    preprocessing_cat = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encode', encoder)])
    
    preprocessing_num = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', scaler)])
    
    
    
    processing = ColumnTransformer(transformers=[
     ('cat', preprocessing_cat, categorical),
     ('num', preprocessing_num, numerical)
     ]
    )
    
    pipe = Pipeline(steps=[('feat', preprocessing_feat), 
                           ('preprocessor', processing),
                           ('clf', clf)], verbose=True)

    return pipe
```

```python
labels = df_set.join(df_labels.set_index('id'))['status_group']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_set, labels, test_size=0.2, random_state=1)
```

```python
categorical_cols, numerical_cols = get_feature_types(df_set)
```

```python
pipe = preprocessing_pipeline(lgb.LGBMClassifier(class_weight={0:0.35, 1:0.5 , 2:0.15}), categorical_cols, numerical_cols)
pipe = pipe.fit(X_train, y_train)
prediction = pipe.predict(X_test)

print('F1 score:', f1_score(y_test, prediction,
                            average='weighted'))


fig, ax = plt.subplots(figsize=(6, 5))
plt.grid(False)
plot_confusion_matrix(pipe, X_test, y_test, display_labels=label_map.keys(), ax=ax) 
```

```python
# Separate columns in numerical and categorical
categorical_cols, numerical_cols = get_feature_types(df_set)

pipe = preprocessing_pipeline_shap(lgb.LGBMClassifier(class_weight={0:0.25, 1:0.5 , 2:0.25}), categorical_cols, numerical_cols)
train = pipe.fit_transform(X_train, y_train)
df_train = pd.DataFrame(train,
            columns = (
                pipe.named_steps['preprocessor'].transformers_[0][1].named_steps['encode'].get_feature_names(categorical_cols).tolist()
                + numerical_cols
            )
        )
```

```python
model = lgb.LGBMClassifier(class_weight={0:0.25, 1:0.5 , 2:0.25})
model.fit(train, y_train)
```

```python
explainer = shap.Explainer(model)
shap_test = explainer(df_train)
```

```python
shap.plots.bar(shap_test[:,:,2])
```
